# Data Visualization
This notebook is used to visualize the data

# a. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# a. Loading data

```python
df = pd.read_csv('../data/processed/cleaned_data.csv')
print('data loaded successfully')
```

# b. Distribution of the feature variables

```python
numeric_features = ["danceability", "energy", "loudness", "speechiness",
                    "acousticness", "instrumentalness", "liveness",
                    "valence", "tempo", "duration_ms"]

# Plot histograms
for col in numeric_features:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()

# For binary features
sns.countplot(x="explicit", data=df)
plt.title("Explicit vs Clean Songs")
plt.show()

sns.countplot(x="mode", data=df)
plt.title("Mode (Major=1, Minor=0)")
plt.show()
```

# c. Distribution of the target variable

```python
plt.figure(figsize=(6,4))
sns.histplot(df["popularity"], bins=30, kde=True)
plt.title(f"Distribution of {"popularity"}")
plt.show()
```
